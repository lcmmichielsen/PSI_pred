## Data distribution

In this notebook, we look at the distribution of the PSI values (Figure 1BC, S1).

Data needed to reproduce the results can be downloaded from Zenodo. 

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns

import matplotlib

matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

### Hippocampus 

First, we load the PSI values. In the histograms we visualize the PSI values of ALL the exons. Since the distribution is quite skewed we downsampled this later on.

```python
PSI_glia = pd.read_csv('../../Zenodo/Human/HPC/PSI/PSI_glia_norm_ALL.csv', index_col=0)
PSI_neur = pd.read_csv('../../Zenodo/Human/HPC/PSI/PSI_neur_norm_ALL.csv', index_col=0)

# Only keep exons that have a PSI value in neurons and glia
idx_tokeep = PSI_glia['0'].notna() & PSI_neur['0'].notna()
PSI_glia = PSI_glia[idx_tokeep]
PSI_neur = PSI_neur[idx_tokeep]

# Define variable exons (delta PSI > 0.25)
PSI_diff = PSI_glia - PSI_neur
PSI_diff['Variable exon'] = np.abs(PSI_diff) > 0.25
```

```python
plt.figure(figsize=[2.5,2.5])
sns.histplot(x = PSI_glia['0'], hue = PSI_diff['Variable exon'], 
             stat='count', multiple='stack', binrange=[0,1],
            binwidth=0.025, palette=['#9DC3E6', '#F4B68C'])
sns.despine()
plt.xlabel('PSI')
plt.ylabel('')
plt.title('Glia (HPC)')
plt.show()

plt.figure(figsize=[2.5,2.5])
sns.histplot(x = PSI_glia['0'], hue = PSI_diff['Variable exon'], 
             stat='count', multiple='stack', binrange=[0,1],
            binwidth=0.025, palette=['#9DC3E6', '#F4B68C'], legend=False)
sns.despine()
plt.xlabel('PSI')
plt.ylabel('')
plt.title('Glia (HPC)')
plt.ylim([0,1000])
plt.show()

plt.figure(figsize=[2.5,2.5])
sns.histplot(x = PSI_neur['0'], hue = PSI_diff['Variable exon'], 
             stat='count', multiple='stack', binrange=[0,1],
            binwidth=0.025, palette=['#9DC3E6', '#F4B68C'])
sns.despine()
plt.xlabel('PSI')
plt.ylabel('')
plt.title('Neurons (HPC)')
plt.show()

plt.figure(figsize=[2.5,2.5])
sns.histplot(x = PSI_neur['0'], hue = PSI_diff['Variable exon'], 
             stat='count', multiple='stack', binrange=[0,1],
            binwidth=0.025, palette=['#9DC3E6', '#F4B68C'], legend=False)
sns.despine()
plt.xlabel('PSI')
plt.ylabel('')
plt.title('Neurons (HPC)')
plt.ylim([0,1000])
plt.show()

plt.figure(figsize=[2.5,2.5])
sns.histplot(x = PSI_diff['0'], hue = PSI_diff['Variable exon'], 
             stat='count', multiple='stack', binrange=[-1,1],
            binwidth=0.05, palette=['#9DC3E6', '#F4B68C'], legend=False)
sns.despine()
plt.xlabel('Delta PSI')
plt.ylabel('')
plt.title('HPC')
plt.xlim([-1,1])
plt.show()

plt.figure(figsize=[2.5,2.5])
sns.histplot(x = PSI_diff['0'], hue = PSI_diff['Variable exon'], 
             stat='count', multiple='stack', binrange=[-1,1],
            binwidth=0.05, palette=['#9DC3E6', '#F4B68C'], legend=False)
sns.despine()
plt.xlabel('Delta PSI')
plt.ylabel('')
plt.title('HPC')
plt.ylim([0,1000])
plt.xlim([-1,1])
plt.show()
```
