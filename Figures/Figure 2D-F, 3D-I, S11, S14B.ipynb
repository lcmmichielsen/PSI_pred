### Plotting RBP binding profiles

Here, we show how to plot the RBP binding profiles. 

The RBP coordinates and PSI values can be downloaded from Zenodo. 

```python
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm

import seaborn as sns
from matplotlib import pyplot as plt

import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42

from plot_peaks_RBP import plot_peaks_RBP
```

### Hippocampus

First, we load the peaks from the eCLIP data. In the dataframe, every row is a gene and every column is an RBP. The values indicate begin and end of peaks measured.

```python
peaks_path = '../../Zenodo/Human/HPC/RBP/RBP_coor_repl.pickle'
RBP_coor_repl, rows, columns = pd.read_pickle(peaks_path)
peaks = pd.DataFrame(RBP_coor_repl, index=rows, columns=columns)
```

Read the PSI values, so we can split later on for exons < 0.5 and exons > 0.5

```python
PSI_glia = pd.read_csv('../../Zenodo/Human/HPC/PSI/PSI_glia_norm.csv', 
                       index_col=0)
PSI_neur = pd.read_csv('../../Zenodo/Human/HPC/PSI/PSI_neur_norm.csv', 
                       index_col=0)
PSI = pd.concat((PSI_glia, PSI_neur), axis=1)
PSI.columns = ['Glia', 'Neurons']

# Only keep the exons with PSI value defined in neurons and glia
PSI = PSI.loc[PSI.notna().sum(axis=1) > 1]
PSI_glia = PSI_glia.loc[PSI.index]
PSI_neur = PSI_neur.loc[PSI.index]
```

Next, we read the RBP counts. This is slightly different from the peaks. Here, there are no coordinates per gene/RBP, but there is counted per exon already how many RBPs bind at a specific location. We use this to filter out the exons with no RBP binding sites since these genes were probably not expressed/measured in the eCLIP data.

```python
RBP_counts = pd.read_csv('../../Zenodo/Human/HPC/RBP/RBP_peaks.csv', 
                         index_col=0)
RBP_counts = RBP_counts.loc[PSI.index]
tokeep = np.sum(RBP_counts, axis=1) > 0
```

Split the names of the exons based so that we get the chromosome, begin, end, gene name and strand

```python
exons_info = pd.DataFrame(RBP_counts.index)[0].str.split('_',expand=True)
exons_info.iloc[:5]
```

Now we can plot the RBPs of interest.

```python
RBPs = ['PTBP1', 'U2AF1', 'U2AF2', 'SF3B4', 'HNRNPC','QKI', 'STAU2', 'HNRNPU']
t = 0.5 #Split between high and low PSI values

for rbp in RBPs:
    plot_peaks_RBP(rbp, exons_info.iloc[tokeep.values], peaks, 
                   PSI_glia[tokeep], PSI_neur[tokeep], 
                   highthres=t, lowthres=t,
                  save=False)
```

### Frontal cortex

```python
peaks_path = '../../Zenodo/Human/FC/RBP/RBP_coor_repl.pickle'
RBP_coor_repl, rows, columns = pd.read_pickle(peaks_path)
peaks_FC = pd.DataFrame(RBP_coor_repl, index=rows, columns=columns)

PSI_glia_FC = pd.read_csv('../../Zenodo/Human/FC/PSI/PSI_glia_norm.csv', 
                       index_col=0)
PSI_neur_FC = pd.read_csv('../../Zenodo/Human/FC/PSI/PSI_neur_norm.csv', 
                       index_col=0)
PSI_FC = pd.concat((PSI_glia_FC, PSI_neur_FC), axis=1)
PSI_FC.columns = ['Glia', 'Neurons']
PSI_FC = PSI_FC.loc[PSI_FC.notna().sum(axis=1) > 1]
PSI_glia_FC = PSI_glia_FC.loc[PSI_FC.index]
PSI_neur_FC = PSI_neur_FC.loc[PSI_FC.index]

RBP_counts_FC = pd.read_csv('../../Zenodo/Human/FC/RBP/RBP_peaks.csv', 
                         index_col=0)

# We have to get rid of the .version in the gene name
xx = pd.DataFrame(RBP_counts_FC.index)[0].str.split('_', expand=True)
RBP_counts_FC.index = xx[0]+'_'+xx[1]+'_'+xx[2]+'_'+xx[3].str.split('.', expand=True)[0]+'_'+xx[4]

RBP_counts_FC = RBP_counts_FC.loc[PSI_FC.index]
tokeep_FC = np.sum(RBP_counts_FC, axis=1) > 0

exons_info_FC = pd.DataFrame(RBP_counts_FC.index)[0].str.split('_',expand=True)
```

```python
### Plots related to Figure 3 - FC

RBPs = ['STAU2', 'HNRNPU']

for rbp in RBPs:
    plot_peaks_RBP(rbp, exons_info_FC.iloc[tokeep_FC.values], peaks_FC, 
                   PSI_glia_FC[tokeep_FC], PSI_neur_FC[tokeep_FC],
                  save=False)
```
